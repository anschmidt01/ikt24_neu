```python
import pandas as pd
from sklearn.cluster import KMeans  # Correct import for KMeans
import ace_tools as tools

# Load the CSV file
file_path = 'Raum639.csv'
df = pd.read_csv(file_path)

# Step 1: Drop columns with no data
df_cleaned = df.drop(columns=["Unnamed: 0", "Unnamed: 1"])

# Step 2: Handle noisy data
# Remove rows where 'count' and 'device_count_consumer' columns have repeating placeholders
df_cleaned = df_cleaned[df_cleaned['count'] != 'count']
df_cleaned = df_cleaned[df_cleaned['device_count_consumer'] != 'device_count_consumer']

# Convert columns to appropriate data types
df_cleaned['count'] = pd.to_numeric(df_cleaned['count'], errors='coerce')
df_cleaned['device_count_consumer'] = pd.to_numeric(df_cleaned['device_count_consumer'], errors='coerce')
df_cleaned['9'] = pd.to_numeric(df_cleaned['9'], errors='coerce')

# Drop any rows that still have NaN values after conversion
df_cleaned.dropna(inplace=True)

# Step 3: Clustering
# Select numerical columns for clustering
numerical_columns = ['8', '9', 'count', 'device_count_consumer']
X = df_cleaned[numerical_columns]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_cleaned['cluster'] = kmeans.fit_predict(X)

# Display the cleaned and clustered data
tools.display_dataframe_to_user(name="Clustered Data", dataframe=df_cleaned)

df_cleaned.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans  # Correct import for KMeans

# Load the CSV file
file_path = 'Raum639.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Step 1: Drop columns with no data
df_cleaned = df.drop(columns=["Unnamed: 0", "Unnamed: 1"])

# Step 2: Handle noisy data
# Remove rows where 'count' and 'device_count_consumer' columns have repeating placeholders
df_cleaned = df_cleaned[df_cleaned['count'] != 'count']
df_cleaned = df_cleaned[df_cleaned['device_count_consumer'] != 'device_count_consumer']

# Convert columns to appropriate data types, filling NaNs with a default value
df_cleaned['count'] = pd.to_numeric(df_cleaned['count'], errors='coerce').fillna(0)
df_cleaned['device_count_consumer'] = pd.to_numeric(df_cleaned['device_count_consumer'], errors='coerce').fillna(0)
df_cleaned['9'] = pd.to_numeric(df_cleaned['9'], errors='coerce').fillna(0)

# Step 3: Clustering
# Select numerical columns for clustering
numerical_columns = ['8', '9', 'count', 'device_count_consumer']
X = df_cleaned[numerical_columns]

# Ensure that we have valid data to cluster
if not X.empty and len(X) > 0:
    # Apply K-Means clustering
    kmeans = KMeans(n_clusters=3, random_state=42)
    df_cleaned['cluster'] = kmeans.fit_predict(X)
    print(df_cleaned.head())  # Display the cleaned and clustered data
else:
    print("No valid data available for clustering.")
```
