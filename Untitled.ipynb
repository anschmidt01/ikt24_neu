```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Bluetooth data
df_bluetooth = pd.read_csv('Raum639.csv')  # Replace with the actual path to your CSV file

# Manually enter the schedule
schedule = {
    'module': ['Datenbank', 'Prog3', 'PGI', 'Prog1', 'DatenSich'],
    'time_start': ['14:00', '15:45', '09:45', '08:00', '14:00'],
    'time_end': ['15:30', '17:15', '11:15', '09:30', '15:30'],
    'date': ['2023-12-11', '2023-12-12', '2023-12-13', '2023-12-14', '2023-12-11']
}

df_schedule = pd.DataFrame(schedule)

# Convert time data into a comparable format
df_bluetooth['timestamp'] = pd.to_datetime(df_bluetooth['2023-12-04T12:24:00Z'])  # Replace with the actual timestamp column name
df_bluetooth['timestamp'] = df_bluetooth['timestamp'].dt.tz_localize(None)  # Remove timezone
df_schedule['datetime_start'] = pd.to_datetime(df_schedule['date'] + ' ' + df_schedule['time_start']).dt.tz_localize(None)
df_schedule['datetime_end'] = pd.to_datetime(df_schedule['date'] + ' ' + df_schedule['time_end']).dt.tz_localize(None)

# Add a 'week' column to group by weeks
df_bluetooth['week'] = df_bluetooth['timestamp'].dt.isocalendar().week

# Initialize an empty list to hold the attendance data for each module, grouped by week
attendance_data_weekly = []

# Iterating over the schedule to calculate weekly attendance
for index, row in df_schedule.iterrows():
    module_attendance = df_bluetooth[
        (df_bluetooth['timestamp'] >= row['datetime_start']) & 
        (df_bluetooth['timestamp'] <= row['datetime_end'])
    ]
    
    # Group by week and sum the device counts during the class time
    weekly_counts = module_attendance.groupby('week')['8'].sum()  # Assuming column '8' holds the device count
    
    for week, count in weekly_counts.items():
        attendance_data_weekly.append({
            'module': row['module'],
            'week': week,
            'attendance_count': count
        })

# Creating a DataFrame with the weekly attendance data
df_attendance_weekly = pd.DataFrame(attendance_data_weekly)

# Plotting the weekly data
plt.figure(figsize=(10, 6))

# Pivot to make the data easier to plot
df_pivot = df_attendance_weekly.pivot(index='week', columns='module', values='attendance_count')

df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Add labels and title
plt.xlabel('Week Number')
plt.ylabel('Number of Attendees')
plt.title('Weekly Attendance per Module')

# Save the plot as a PNG file in the current working directory
plt.tight_layout()
plt.savefig("woechentlichxy_abbildung.png")

# Optionally, you can display the plot as well
plt.show()
```
